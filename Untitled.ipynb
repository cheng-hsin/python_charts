```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取檔案
plt.rcParams['font.sans-serif'] = 'Microsoft JhengHei'
df = pd.read_csv("salary.csv", encoding= 'utf-8')
```

```python
# 取出參數
xyear = df["Year"]
```

```python
print(df)
```

```python
#color=["tab:brown","tab:green","tab:red","tab:blue","tab:orange"]
pop=[df["臺灣地區"],df["新北市"],df["臺北市"],df["桃園市"],df["臺中市"]
     ,df["臺南市"],df["高雄市"],df["宜蘭縣"],df["新竹縣"],df["苗栗縣"]
     ,df["彰化縣"],df["南投縣"],df["雲林縣"],df["嘉義縣"],df["屏東縣"]
     ,df["臺東縣"],df["花蓮縣"],df["澎湖縣"],df["基隆市"],df["新竹市"]
     ,df["嘉義市"]]

lbl=['臺灣地區','新北市','臺北市','桃園市','臺中市','臺南市','高雄市'
     ,'宜蘭縣','新竹縣','苗栗縣','彰化縣','南投縣','雲林縣','嘉義縣'
     ,'屏東縣','臺東縣','花蓮縣','澎湖縣','基隆市','新竹市','嘉義市']

# 全線圖
plt.figure(figsize=(10,8),dpi=100)
for i in range(0,21):
     plt.plot(xyear, pop[i], label=lbl[i])
plt.title("各縣市別平均每戶所得收入總計(1998年起)")
plt.ylabel("年薪")
plt.legend(loc='upper right')
plt.show()
#plt.savefig("臺北捷運各路線年度進出站人次圖_全線圖.png")
```
